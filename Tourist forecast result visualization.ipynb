```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Singapore_Tourist_Forecast_Result.csv")
data.head()
```

```python
data.info()
```

```python
#time_series_timestamp vẫn còn ở dạng obj -> chuyển sang dạng datetime
data['time_series_timestamp'] = pd.to_datetime(data['time_series_timestamp'])

#visual kết quả dựa vào 2 cột time_series_timestamp và time_series_data
plt.figure(figsize=(14, 7))
sns.lineplot(
    x='time_series_timestamp',
    y='time_series_data',
    hue='time_series_type',
    data=data
)
plt.title('Singapore Tourist Over Time')
plt.xlabel('Month')
plt.ylabel('Tourist')
plt.grid(True)
plt.show()
```

Kết luận: Kết quả dự báo cho thấy số lượng khách du lịch Singapore có xu hướng tăng trong 12 tháng tiếp theo và có yếu tố tăng giảm theo mùa vụ.
