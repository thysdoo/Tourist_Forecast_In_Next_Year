```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Singapore_Tourist_Forecast_Result.csv")
data.head()
```

```python
data.info()
```

```python
#time_series_timestamp vn cn  dng obj -> chuyn sang dng datetime
data['time_series_timestamp'] = pd.to_datetime(data['time_series_timestamp'])

#visual kt qu da vo 2 ct time_series_timestamp v time_series_data
plt.figure(figsize=(14, 7))
sns.lineplot(
    x='time_series_timestamp',
    y='time_series_data',
    hue='time_series_type',
    data=data
)
plt.title('Singapore Tourist Over Time')
plt.xlabel('Month')
plt.ylabel('Tourist')
plt.grid(True)
plt.show()
```
